## Analysis of 3808 companies between 2014 and 2018
### A quick analysis of picking stocks based on net profit margin
---

```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
```

   #### Read financial data between 2014 and 2018. Source: [Kaggle dataset](https://www.kaggle.com/datasets/cnic92/200-financial-indicators-of-us-stocks-20142018?resource=download)
---

```python
df = []
for i in range(2014, 2019):
    df.append(pd.read_csv(f"{i}_Financial_Data.csv", index_col=0))
df[0]
```

```python
profits = pd.DataFrame()
for i in range(0, 5):
    profits[str(2014+i)] = df[i]['Gross Profit']
profits
```

   ### Get all columns containing data on profit
---

```python
profit_cols = list(filter(lambda x: "Profit" in x, df[0].columns.to_list()))
profit_cols
```

```python
profit_margin = pd.DataFrame()
for i in range(0, 5):
    profit_margin[str(2014+i)] = df[i]['netProfitMargin']
profit_margin = profit_margin.transpose()
profit_margin+=1
profit_margin
```

```python
pm=profit_margin[profit_margin[profit_margin.columns] > 0].dropna(axis=1)
pm
```

```python
pm=pm.cumprod()
res = pm.loc['2018']
```

```python
res[res>0].sort_values().plot()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
res[res>0].plot()
plt.show()
```

```python
plt.figure()
plt.bar(res[res.values > 25].index, res[res.values > 25].values)
```
